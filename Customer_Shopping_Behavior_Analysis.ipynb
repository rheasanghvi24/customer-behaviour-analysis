```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.xls')
```

```python
df.head()
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
# 1. Top spending age group
print(df.groupby('age_group')['purchase_amount'].mean().round(2).sort_values(ascending=False))

print()

# 2. Discount usage %
print("Discount users:", round(df['discount_applied'].value_counts(normalize=True)['Yes'] * 100, 2), "%")

print()

# 3. Top category by revenue
print(df.groupby('category')['purchase_amount'].sum().round(2).sort_values(ascending=False).head(5))

print()

# 4. Top payment method
print(df['payment_method'].value_counts(normalize=True).mul(100).round(2).head(3))
```
